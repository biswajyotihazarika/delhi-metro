```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("Delhi metro.csv")

# Clean column names
df.columns = df.columns.str.strip()

# Convert Opened(Year) column to datetime
df['Opened(Year)'] = pd.to_datetime(df['Opened(Year)'], errors='coerce')

# Count number of stations per metro line
line_counts = df['Metro Line'].value_counts()

# Plot bar chart
plt.figure(figsize=(10, 6))
plt.bar(line_counts.index, line_counts.values, color='steelblue')
plt.title("Number of Stations per Metro Line")
plt.ylabel("Number of Stations")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("Delhi metro.csv")

# Clean column names
df.columns = df.columns.str.strip()

# Convert Opened(Year) column to datetime
df['Opened(Year)'] = pd.to_datetime(df['Opened(Year)'], errors='coerce')

# Compute average distance between stations in each line
avg_dist_per_line = df.groupby('Metro Line')['Dist. From First Station(km)'].apply(
    lambda x: x.sort_values().diff().mean()
)

# Print results
print("\nAverage Inter-Station Distance by Line (in km):")
print(avg_dist_per_line.round(2))
```
